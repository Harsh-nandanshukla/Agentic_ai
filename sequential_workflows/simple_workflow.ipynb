```python
from langgraph.graph import StateGraph,START,END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatOpenAI()
```

```python
# craet a state
class qna(TypedDict):
    query: str
    answer: str
```

```python
def qst_ans(state:qna)->qna:
    # extract question
    qst=state['query']
    # create a prompt
    prompt=f'Answer the following {qst}'
    # call the model and sentd the prompt 
    answer=model.invoke(prompt).content
    # puth th naswer by llm uinto answer state 
    state['answer']=answer
    return state
```

```python
# crated a grapgh
graph=StateGraph(qna)
# crate a node
graph.add_node('llm_qna',qst_ans)
# create edges
graph.add_edge(START,'llm_qna')
graph.add_edge('llm_qna',END)
# complie the graphs
workflow=graph.compile()
```

```python
workflow.invoke({'query':'What is black hole in 50 words?'})
```
